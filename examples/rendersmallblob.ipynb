```python
import FreeCAD
import MeshPart
doc = FreeCAD.open("smallblob.FCStd")

```

```python
doc.Objects
```

```python
b = doc.Objects[-1]
mesh = MeshPart.meshFromShape(b.Shape)

print(mesh.CountEdges)
print(mesh.CountFacets)
print(mesh.CountPoints)
```

```python
print(len(mesh.Points), mesh.Points[0])
print(len(mesh.Facets), mesh.Facets[0])
```

```python
mesh.Points[50]
```

```python
X = [p.x  for p in mesh.Points]
Y = [p.y  for p in mesh.Points]
Z = [p.z  for p in mesh.Points]
tris = [f.PointIndices  for f in mesh.Facets]
```

```python
import ipyvolume as ipv

ipv.figure()
ipv.xyzlim(-40,40)
mesh = ipv.plot_trisurf(X, Y, Z, triangles=tris, color='orange')
ipv.show()
```

```python
dir(b.Shape)
points, tris = b.Shape.tessellate(0.02)

X = [p.x  for p in points]
Y = [p.y  for p in points]
Z = [p.z  for p in points]
```

```python
ipv.figure()
ipv.xyzlim(-40,40)
mesh = ipv.plot_trisurf(X, Y, Z, triangles=tris, color='orange')
ipv.show()
```
